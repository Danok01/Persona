```python
import matplotlib.pyplot as plt
import numpy as np

miles = df['Miles']
income = df['Income']

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(miles, income, alpha=0.7, label='Data')  # Adjust alpha for transparency

# Calculate and plot the regression line
m, b = np.polyfit(miles, income, 1)  # Linear regression (degree 1)
plt.plot(miles, m * miles + b, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('Miles')
plt.ylabel('Income')
plt.title('Miles vs. Income')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Customers from 150 Miles amd above are considered to be wealthier than others
```

```python
# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Age'], df['Miles'], color='skyblue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Miles Driven')
plt.title('Miles Driven by Age')

# Show the plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# This chart shows people in early 30s to early 20s ae more likely to buy expensive products
```

```python
# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Product'], df['Income'], color='skyblue')

# Add labels and title
plt.xlabel('Product')
plt.ylabel('Income')
plt.title('Product attracted by income')

# Show the plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
